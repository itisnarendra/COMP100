### Creating Tuples

#### 1: Empty constructor
Create an empty tuple and print its type.

#### 2: Tuples by literal assignment
Create a tuple with the values `42`, `'Python'`, `3.14`, `True` and print it and verify the type.

#### 3: Single item tuple
Create a single-element tuple containing only the number 7. Print it and verify its type.

#### 4: Accessing Elements
Given the tuple:   
`t = ('red', 'green', 'blue', 'yellow', 'black')`  
-   Print the first and last element.
-   Print a slice that contains only ('green', 'blue').
-   Print the length of the tuple.

#### 5: Unpacking
Unpack the tuple `(10, 20, 30)` into three variables `x`, `y`, `z`. Print them.

#### 6: Unpacking
Use the star operator to unpack the tuple `(1, 2, 3, 4, 5)` so that the first element goes into `a`, and the rest into `b`. Print both.

#### 7: Tuple Methods
Given:  
`nums = (4, 2, 7, 2, 9, 2, 5)`  
-   Count how many times 2 appears.
-   Find the index of the first occurrence of 7.

### 8: Challenge - Working with Nested Tuples 
Given: 
```
matrix = (    
    (1, 2, 3),
    (4, 5, 6),
    (7, 8, 9) 
)             
```    
-   Print the element in the second row, third column (should be 6).
-   Write a loop to print each row on a separate line.

#### 11


#### 12
