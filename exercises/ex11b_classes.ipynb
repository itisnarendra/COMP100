## Instructor Class   
Create a class that represents a college instructor.   
This is similar to the previous class that you built but with the following changes: constructor takes an additional argument, the `greet` method is replaced by the `__ str __` and two additional methods are added.

---   

### Class Name
`Instructor`

---
#### Constructor (`__ init __`)
The constructor should:
-   Accept three parameters:
    -   `name` - the instructor's name
    -   `department` - the department they belong to
    -   `courses` - a list of the courses taught by this instructor (optional: default value should be `None`)
-   Store `name` and `department` in instance variables
-   for the `courses` parameter:
    -   If it is `None` store an empty list
    -   Otherwise, store the provided list

---
#### Instance Method: (`__str__`)
-   Takes no additional arguments.
-   Returns a string introducing the instructor in the format:   
Hello, my name is &lt;name&gt; and I teach &lt;courses&gt; in &lt;department&gt;.   
*Hint: Use an f-string to build the return value*

---
#### Instance Method: `add_course`
-   Takes one parameter:
    -   `course` - a string that represent the course name
-   Appends it to the instructor's course list
-   Returns nothing

---
#### Instance Method: `remove_course`
-   Takes one parameter:
    -   `course` - a string that represent a course name
-   If the course exist in the instructor's list remove it
-   Returns nothing

---
##### Test harness
```python
print(f'\nCreating first object')
ilia = Instructor('Ilia', 'ICET', 'Python, Data Science'.split(', '))
print(ilia)
course = 'Machine Learning'
print(f'Adding course: {course}')
ilia.add_course(course)
print(ilia)

print(f'\nCreating second object')
arben = Instructor('Arben', 'ICET', ['Java', 'Databases'])
print(arben)
course = 'Java'
print(f'Removing course: {course}')
arben.remove_course(course)
print(arben)
```

---
##### Expected output
Creating first object                  
Hello, my name is Ilia and I teach ['Python', 'Data Science'] in ICET.                  
Adding course: Machine Learning                  
Hello, my name is Ilia and I teach ['Python', 'Data Science', 'Machine Learning'] in ICET.                  

Creating second object                  
Hello, my name is Arben and I teach ['Java', 'Databases'] in ICET.                  
Removing course: Java                  
Hello, my name is Arben and I teach ['Databases'] in ICET.

Create a class that represents a grocery shopping cart.   
The class should include a **class attribute**: a dictionary that stores the prices of each items.   

---   

### Class Name
`ShippingCart`

---
#### Class Attribute (`prices`)
A dictionary mapping item names (strings) to their unit price (float).   
You may use the following definition:
```
prices = {'apple': 2.5, 'banana': 1.3, 'orange': 1.8}
```    
This is used by the `calculate_cart_price` method to calculate the price of all the items in the cart

---
#### Constructor (`__ init __`)
The constructor should:
-   Takes one additional parameter:
    -   `name` - the shopper's name
-   Store `name` in an instance variable
-   Create an instance variable to store the shopper's items
    -   This should be an empty list at the start.

---
#### Instance Method: (`__str__`)
-   Takes no additional arguments.
-   Returns a string introducing the shopper in the format:   
&lt;name&gt;'s cart: &lt;items&gt; total prince: &lt;amount&gt;.      
Where:    
-   <items> is the current list of (item, quantity) tuples.
-   <amount> is the total cost formatted to 2 decimal places.

*Hint: Use an f-string to build the return value   
The amount is calculated by calling the `calculate_cart_price()` method.*

---
#### Instance Method: `add_item`
-   Takes two additional parameters:
    -   `item` - a string that represent the item name
    -   `quantity` - a number that represent how many units of the item will be purchased (optional; default is 1)
-   Appends a tuple of the item, quantity to the shopper's item list
-   Returns nothing

---
#### Instance Method: `remove_item`
-   Takes one parameter:
    -   `item` - a string that represent name of the item to remove from the cart
-   Checks if the item exist in the shopper's list of courses
-   Returns nothing

---
#### Instance Method: `calculate_cart_price`
-   Takes no additional arguments.
-   Gets the price of each `item` from the class variable `prices`. (Do not use the key to access the price, instead use the `get()` method of the dictionay class to return a prive of `1` if the key is not in the dictionary.)
    -   it multiples the quantity by the price.
    -   it sum of all of the above product to get the `total price`
-   Returns `total price`
-   This method is used in the `__ str __` method to return the price

---
##### Test harness
```python
cart = ShoppingCart('Narendra')
print(f'{cart}\n')

item, qty = 'apple', 3
print(f'Adding {qty} {item} to cart')
cart.add_item(item, qty)

item, qty = 'banana', 5
print(f'Adding {qty} {item} to cart')
cart.add_item(item, qty)

item, qty = 'pear', 2
print(f'Adding {qty} {item} to cart')
cart.add_item(item, qty)
print(f'{cart}\n')

item = 'apple'
print(f'Removing {item} from the cart')
cart.remove_item('apple')
print(f'{cart}\n')
```
